```python
from nipype.interfaces.base import (
    BaseInterfaceInputSpec,
    File,
    SimpleInterface,
    TraitedSpec,
    isdefined,
    traits,
)
```

```python
class Dog:
    def __init__(self, a1, a2):
        self.a1 = a1
        self.a2 = a2

class TESTInputSpec(BaseInterfaceInputSpec):
    a1 = traits.Str()
    a2 = traits.Float()

class TESTOutputSpec(TraitedSpec):
    pass

class TEST(SimpleInterface):
    input_spec = TESTInputSpec
    output_spec = TESTOutputSpec

    def _run_interface(self, runtime):
        dog = Dog(a1=self.inputs.a1, a2=self.inputs.a2)
        self._used_params = [k for k in self.inputs.get().keys() if hasattr(dog, k)]
        return runtime

    def _list_outputs(self):
        inputs = self.inputs.get()
        # Convert _Undefined values to "n/a"
        inputs = {k: "n/a" if not isdefined(v) else v for k, v in inputs.items()}
        
        model_params = {
            "a1": "RadialOrder",
            "a2": "LaplacianRegularization",
        }
        model_params = {
            v: inputs.get(k, "n/a") for k, v in model_params.items() if k in self._used_params
        }
        print(model_params)
```

```python
v = TEST(a1="hi", a2=4)

v.run()
```
